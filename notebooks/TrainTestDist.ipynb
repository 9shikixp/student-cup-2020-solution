基本はフォーラムへの投稿を見て，train test のカテゴリ割合の違いを考えるように．  
もう一つの根拠として，トピックモデルからも train test の違いを推察できるかもという話．

```python
from collections import Counter
import pandas as pd
```

```python
train_df = pd.read_csv('../processed_data/train_df_2865_topic_theta.csv')
test_df = pd.read_csv('../processed_data/test_df_topic_theta.csv')
```

```python
topic_category_df = pd.DataFrame()
for topic_id in range(20):
    topic_category_df[topic_id] = (train_df[train_df['topic_id'] == topic_id].jobflag.value_counts())
topic_category_df.sort_index(inplace=True)
```

### トピックidごとに，ある程度分かれている

```python
topic_category_df.T
```

```python
topic_dist_df = pd.concat([train_df.topic_id.value_counts(normalize=True), test_df.topic_id.value_counts(normalize=True)], axis=1)
topic_dist_df.columns = ['train', 'test']
topic_dist_df['diff'] = topic_dist_df['test'] - topic_dist_df['train']
```

### trainとtestでトピックの分布が異なる

```python
topic_dist_df.sort_values('diff')
```
